# Traitement du fichier pour l'année 2013 en vu de le rendre compatible avec le fichier 2015

```python
import pandas as pd
import numpy as np
```

```python
df_2013 = pd.read_csv('./Car_CO2_Emissions/cl_JUIN_2013-complet3.csv', encoding='ANSI', sep=';')
df_2013.head()
```

```python
df_2013.columns
```

```python
df_2013_cleaned_gamme = df_2013.drop(['Modèle UTAC', 'Désignation commerciale', 'CNIT', 'Type Variante Version (TVV)', 'HC (g/km)', 'HC+NOX (g/km)', 'masse vide euro max (kg)', 'Date de mise à jour'], axis=1)
```

```python
df_2013_cleaned_gamme.info()
```

```python
df_2013_cleaned_gamme.head()
```

```python
df_2013_cleaned = df_2013_cleaned_gamme.drop(['Carrosserie', 'gamme'], axis=1)
df_2013_cleaned.head(2)
```

```python
df_2013_cleaned.info()
```

```python
df_2013_without_na = df_2013_cleaned.dropna()
df_2013_without_na.shape
```

```python
df_2013_without_na['Boîte de vitesse'].value_counts()
```

```python
df_2013_without_na['nb_rapp'] = df_2013_without_na['Boîte de vitesse']
```

```python
df_2013_without_na = df_2013_without_na.drop(df_2013_without_na[(df_2013_without_na['Boîte de vitesse'] == 'V 0') | (df_2013_without_na['Boîte de vitesse'] == 'S 6')].index)
```

```python
df_2013_without_na['Boîte de vitesse'][df_2013_without_na['Boîte de vitesse'].str.contains("A")] = "A"
df_2013_without_na['Boîte de vitesse'][df_2013_without_na['Boîte de vitesse'] != 'A'] = 'M'
```

```python
df_2013_without_na['Boîte de vitesse'].value_counts()
```

```python
df_2013_without_na.nb_rapp = df_2013_without_na.nb_rapp.apply(lambda x : x[2])
```

```python
df_2013_without_na.head()
```

```python
cols = df_2013_without_na.columns.tolist()
cols = [cols[1], cols[0]] + cols[2:]
df_2013_without_na = df_2013_without_na[cols]
```

```python
df_2013_without_na.columns = ['modele', 'marque', 'energie', 'hybride', 'puiss_admin',
       'p_kw', 'typ_boite', 'conso_urbaine', 'conso_autoroute',
       'conso_mixte', 'co2', 'co', 'nox', 'part_fine',
       'poids', 'norme', 'nb_rapp']
```

### Application des mêmes transformations que pour la liste 2015 

```python
df_2013_without_na.loc[df_2013_without_na['norme'].str.contains('EURO5'), 'norme'] = '0'
df_2013_without_na.loc[df_2013_without_na['norme'].str.contains('EURO6'), 'norme'] = '1'
df_2013_without_na.loc[df_2013_without_na['typ_boite'].str.contains('M'), 'typ_boite'] = '0'
df_2013_without_na.loc[df_2013_without_na['typ_boite'].str.contains('A'), 'typ_boite'] = '1'
df_2013_without_na.loc[df_2013_without_na['hybride'].str.contains('non'), 'hybride'] = '0'
df_2013_without_na.loc[df_2013_without_na['hybride'].str.contains('oui'), 'hybride'] = '1'
df_2013_without_na.loc[df_2013_without_na['energie'].str.contains('ES/GN '), 'energie'] = 'ES'
df_2013_without_na.loc[df_2013_without_na['energie'].str.contains('ES '), 'energie'] = 'ES'
df_2013_without_na.norme = pd.to_numeric(df_2013_without_na.norme, errors='ignore')
df_2013_without_na.nb_rapp = pd.to_numeric(df_2013_without_na.nb_rapp, errors='ignore')
df_2013_without_na.conso_urbaine = pd.to_numeric(df_2013_without_na.conso_urbaine, errors='ignore')
df_2013_without_na.conso_autoroute = pd.to_numeric(df_2013_without_na.conso_autoroute, errors='ignore')
df_2013_without_na.puiss_admin = pd.to_numeric(df_2013_without_na.puiss_admin, errors='ignore')
df_2013_without_na.p_kw = pd.to_numeric(df_2013_without_na.p_kw, errors='ignore')
df_2013_without_na.hybride = pd.to_numeric(df_2013_without_na.hybride, errors='ignore')
df_2013_without_na.typ_boite = pd.to_numeric(df_2013_without_na.typ_boite, errors='ignore')
```

```python
df_2013_without_duplicates = df_2013_without_na.drop_duplicates()
```

```python
df_2013_without_duplicates.shape
```

```python
df_2013_without_duplicates.head()
```

```python
df_2013_without_duplicates.to_csv('cleaned_2013.csv', index=False)
```
